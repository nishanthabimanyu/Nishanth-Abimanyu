```python
# ======================================================================
# Step 1: Install elan (Lean's toolchain manager)
# ======================================================================
!curl https://raw.githubusercontent.com/leanprover/elan/master/elan-init.sh -sSf | sh -s -- -y

# ======================================================================
# Step 2: Add elan to the system's PATH for this session
# ======================================================================
import os
os.environ['PATH'] = f"/root/.elan/bin:{os.environ['PATH']}"

# ======================================================================
# Step 3: Install the lean-dojo Python library
# ======================================================================
!pip install lean-dojo

# ======================================================================
# Step 4: Run the LeanDojo example code
# ======================================================================
from lean_dojo import LeanGitRepo, Theorem, Dojo, ProofFinished

# 1. Define the Lean repository we want to work with.
repo = LeanGitRepo(
    "https://github.com/yangky11/lean4-example",
    "7b6ecb9ad4829e4e73600a3329baeb3b5df8d23f"
)

# 2. Define the specific theorem we want to prove.
theorem = Theorem(repo, "Lean4Example.lean", "hello_world")

# 3. Create a Dojo environment for the theorem.
# This will automatically trace the repo if it hasn't been traced before.
with Dojo(theorem) as (dojo, initial_state):
    print("--- Initial Proof State ---")
    print(initial_state.pp)

    # 4. Run a tactic to advance the proof.
    result = dojo.run_tac(initial_state, "rw [add_assoc, add_comm b, ←add_assoc]")

    print("\n--- Result after running tactic ---")
    print(result)

    # Check if the proof is finished.
    if isinstance(result, ProofFinished):
        print("\nProof finished successfully! 🎉")
```

```python
from lean_dojo import *

repo = LeanGitRepo("https://github.com/yangky11/lean4-example", "7b6ecb9ad4829e4e73600a3329baeb3b5df8d23f")
theorem = Theorem(repo, "Lean4Example.lean", "hello_world")

with Dojo(theorem) as (dojo, initial_state):
    print("--- Initial State ---")
    print(initial_state.pp)

    # First tactic: apply associativity
    state_1 = dojo.run_tac(initial_state, "rw [add_assoc]")
    print("\n--- After `rw [add_assoc]` ---")
    print(state_1.pp)

    # Second tactic: apply commutativity
    state_2 = dojo.run_tac(state_1, "rw [add_comm b]")
    print("\n--- After `rw [add_comm b]` ---")
    print(state_2.pp)

    # Final tactic: apply associativity again
    result = dojo.run_tac(state_2, "rw [←add_assoc]")
    print("\n--- Final Result ---")
    print(result)

    if dojo.is_successful:
        print("\nProof finished successfully!")
```
